# **Missing Value**

```python
mydata.describe(include = 'all')
```

```python
#display all the columns

pd.set_option('display.max_columns', None)

mydata.head(10)
```

```python
#check missing value
#False - no missing value
#True - has missing value

mydata.isnull()
```

```python
mydata.isnull().sum() #to calculate how many missing values
```

```python
#replace the '?' by NaN numpy
#sometimes ? is seen as string by pandas
#Hence, should call numpy to read the NaN

import numpy as np

mydata.replace('?', np.nan, inplace = True)

mydata.isnull().sum()
```

```python
# ? changed to NaN
mydata.head(10)
```

```python
#Identified missing value
# example: normalized-class ; 205 - 164 = missing value
mydata.info()
```

```python
#numerical data = replaced with mean value
#categorical data = replaced with the most preferred one

mean_norm = mydata['normalized-losses'].mean()

mean_norm
```

```python
#cannot run because the datatype is object
#Hence, change the datatype to integer
# Better to convert float to maintain accuracy (safest)
# if convert from float (original data) to integer, probably lose the data - accuracy

mean_norm = mydata['normalized-losses'].astype('float').mean()

mean_norm
```

```python
#replace the value as missing value

mydata['normalized-losses'].replace(np.nan, mean_norm, inplace = True)

mydata.isnull().sum()
```

```python
#for categorical data, we have to find out the most frequent to replace for the missing value
#top = most freq
#unique = number of categories
mydata.describe(include ='all')
```

```python
#it will count in ascending order and give output for the most used category
mydata['num-of-doors'].value_counts()
```

```python
mydata['num-of-doors'].value_counts().idxmax() #to identify top frequency category
```

```python
mydata['num-of-doors'].replace(np.nan,mydata['num-of-doors'].value_counts().idxmax(),inplace = True)

mydata.isnull().sum()
```

```python
#calculating mean value for 'bore' column
mean_bore = mydata['bore'].astype('float').mean()

mean_bore
```

```python
mydata['bore'].replace(np.nan, mean_bore, inplace = True)

mydata.isnull().sum()
```

```python
mean_stroke = mydata['stroke'].astype('float').mean()
mean_horse = mydata['horsepower'].astype('float').mean()
mean_peak = mydata['peak-rpm'].astype('float').mean()

mydata['stroke'].replace(np.nan, mean_stroke, inplace = True)
mydata['horsepower'].replace(np.nan, mean_horse, inplace = True)
mydata['peak-rpm'].replace(np.nan, mean_peak, inplace = True)

mydata.isnull().sum()

```

```python
#dropping missing value for price

mydata.dropna(subset = ['price'], axis = 0, inplace = True) #axis = 0 because we are only dropping the rows of missing value

mydata.isnull().sum()
```

# **Data Normalisation**

*   Simple Feature Scalling
*   Min-Maz

*   Zscore







```python
#simple feature scalling
#single sequel bracket if 1 column, double [[]] if more than 1
mydata['width'].head()
```

```python
mydata[['width','length','height']]. head()
```

```python
#simple feature scalling method for 'width'

mydata['width'] = mydata['width']/mydata['width'].max()

mydata[['width','length','height']]. head()
```

```python
#min and max value changed
mydata['width'].describe()
```

```python
#Min-Max method for 'width'

mydata['length'] = (mydata['length'] - mydata['length'].min()) / (mydata['length'].max() - mydata['length'].min())



mydata[['width','length','height']]. head()
```

```python
mydata['length'].describe()
```

```python
#Zscore method

mydata['height'] = (mydata['height'] - mydata['height'].mean()) / mydata['height'].std()

mydata[['width','length','height']].head()
```

```python
mydata[['width','height','length']].describe()
```

# **Data Binning**

```python
#categorising the price crange

mydata['price'].describe()
```

```python
price_bins = np.linspace(mydata_new1['price'].min(), mydata_new1['price'].max(), num = 4)

print(price_bins)

price_group = ['Low Price', 'Mid Price', 'High Price']

mydata_new1['price_bins'] = pd.cut(mydata_new1['price'], price_bins, labels = price_group, include_lowest = True)

mydata_new1[['price','price_bins']].head()
```

```python
#there is much difference between low price and the rest. - class imbalance
#there is high chance for the model to predict mistakenly the low price rather than the rest
mydata_new1['price_bins'].value_counts()
```

```python
mydata_new1.describe() #only to see the numerical data, so no need to include = 'all'
```

```python
norm_bins = np.linspace(mydata_new1['normalized-losses'].min(), mydata_new1['normalized-losses'].max(), num = 4)

print(norm_bins)

norm_group = ['Low', 'Mid', 'High']

mydata_new1['normalized-losses_bins'] = pd.cut(mydata_new1['normalized-losses'], norm_bins, labels = norm_group, include_lowest = True)

mydata_new1[['normalized-losses','normalized-losses_bins']].head(20)
```
