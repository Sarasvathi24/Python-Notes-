# **Creating Own Dataset using Numpy**

```python
# pandas change the dictionary to dataframe
# toy dataset = small dataset
# creating numerical data

import pandas as pd
import numpy as np
from datetime import datetime, timedelta #dealing with date and time

#create numerical data

col1 = np.random.normal(250,10,500) #(mean/average,standard deviation,total number of rows)
col2 = np.random.normal(100,20,500)
col3 = np.random.normal(750,5,500)

#creating categorical data

categorical_data = ['A','B','C','D','E','F','G','H','I','J']

col4 = np.random.choice(categorical_data,500)
col5 = np.random.choice(categorical_data,500)

#creating date data

start_date = datetime(2022,1,1)
end_date = datetime(2024,1,1)

delta = end_date - start_date
date_list = [start_date + timedelta(days = x) for x in range(delta.days)]

col6 = np.random.choice(date_list,500)

#combine all the columns/features

data = np.column_stack((col1,col2,col3,col4,col5,col6))
#changing them into dataframe
x = pd.DataFrame(data,columns = ['Num1','Num2','Num3','Cat1','Cat2','Date']) #labelling the columns

x.head()
```

```python
x.shape
```

```python
x.info()
```

```python
x.describe()
```

```python
#unique = number of the categories in the categorical column (appears most often)
#freq = frequency of top
x.describe(include ='all')
```

```python
#changing the datatype of Num1 from object to float
#x[['Initial name of column1','Initial name of column2']] = x[['Initial name of column1','Initial name of column2']].astype('float')

x['Num1'] = x['Num1'].astype('float')

x[['Num2','Num3']] = x[['Num2','Num3']].astype('float') #changing 2 columns to float

x.dtypes
```

```python
x.describe() #doesn't include categorical data
```

```python
x.describe(include = 'all')
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta #dealing with date and time

#create numerical data

column1 = np.random.normal(25,1,50)
column2 = np.random.normal(30,2,50)
column3 = np.random.normal(40,7,50)

#creating categorical data

categorical = ['Action','Thriller','Crime','Comedy','Crime Thriller','Romance']

column4 = np.random.choice(categorical,50)
column5 = np.random.choice(categorical,50)

#creating date data

StartDate = datetime(2020,1,1)
EndDate = datetime(2024,12,31)

a = EndDate - StartDate
date_line = [StartDate + timedelta(days = x) for x in range(a.days)]

column6 = np.random.choice(date_line,50)

#combine all the columns/features

data1 = np.column_stack((column1,column2,column3,column4,column5,column6))
#changing them into dataframe
y = pd.DataFrame(data1,columns = ['Age - Female','Age - Male','Age - Unknown','Genre1','Genre2','Date']) #labelling the columns

y.head()
```

```python
y.describe(include = 'all')
```
